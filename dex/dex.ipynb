```python
import pandas as pd
import requests

# 设置 API URL
#coin_id = 'pudgy-penguins'
coin_id = 'bonk'
url = f"https://pro-api.coingecko.com/api/v3/coins/{coin_id}"

headers = {
    'Accept': 'application/json',
    'X-CG-PRO-API-KEY': api_key,
}

# 调用 API
response = requests.get(url, headers=headers)

if response.status_code == 200:
    # 将响应数据转换为 DataFrame
    data = response.json()
    df = pd.json_normalize(data)
    
    # 查看 DataFrame
    #print(df)
else:
    print(f"Error fetching data: {response.status_code}")

df
```

```python
# 找到以 'platforms.' 开头的列
platform_columns = [col for col in df.columns if col.startswith('platforms.')]

# 创建一个新的 DataFrame
new_rows = []

# 提取数据
for col in platform_columns:
    chain_name = col.replace('platforms.', '')  # 获取 'platforms.' 后面的部分
    # 创建每一行的字典
    new_rows.extend([{'chain': chain_name, 'address': address} for address in df[col]])

# 转换为 DataFrame
new_df = pd.DataFrame(new_rows)

# 查看结果
new_df
```

```python
import os
import requests
import pandas as pd
from dotenv import load_dotenv

# 加载 .env 文件中的环境变量
load_dotenv()

# 获取 API 密钥
api_key = os.getenv('API_KEY')

network = 'solana'
token_address = 'DezXAZ8z7PnrnRJjz3wXBoRgixCa6xjnB7YaB1pPB263'
page = 1  # 初始页数
all_data = []  # 存储所有数据

# API URL
url = f'https://pro-api.coingecko.com/api/v3/onchain/networks/{network}/tokens/{token_address}/pools?page='

# 设置请求头
headers = {
    'Accept': 'application/json',
    'X-CG-PRO-API-KEY': api_key,
}

# 无限循环遍历
while True:
    response = requests.get(url + str(page), headers=headers)

    # 检查请求是否成功
    if response.status_code == 200:
        data = response.json()
        print(f"Response for page {page}: {data}")  # 输出返回的原始数据

        # 检查 'data' 键是否存在且非空
        if 'data' in data and isinstance(data['data'], list) and len(data['data']) > 0:
            all_data.extend(data['data'])  # 将数据添加到 all_data
            page += 1  # 继续到下一页
        else:
            print(f"No more data found at page {page}. Stopping.")
            break  # 如果没有数据，停止循环
    else:
        print(f"Error on page {page}: {response.status_code} - {response.text}")
        break  # 请求失败时停止循环

# 将数据保存到 CSV 文件
if all_data:
    df = pd.DataFrame(all_data)
    df.to_csv('output_data.csv', index=False)
    print("Data has been saved to output_data.csv")
else:
    print("No data to save.")
```

```python
import pandas as pd
import requests
import os

# 清空 ohlcv_data 目录
output_dir = 'ohlcv_data'
if os.path.exists(output_dir):
    for file in os.listdir(output_dir):
        os.remove(os.path.join(output_dir, file))
else:
    os.makedirs(output_dir)

filtered_rows['network'] = filtered_rows['id'].apply(lambda x: x.split('_')[0])
filtered_rows['pool_address'] = filtered_rows['id'].apply(lambda x: x.split('_')[1])

output_dir = 'ohlcv_data'
os.makedirs(output_dir, exist_ok=True)

base_url = "https://pro-api.coingecko.com/api/v3/onchain/networks/{network}/pools/{pool_address}/ohlcv/{timeframe}?limit=1000&before_timestamp=1732982400"

headers = {
    'Accept': 'application/json',
    'X-CG-PRO-API-KEY': api_key,
}

for index, row in filtered_rows.iterrows():
    network = row['network']
    pool_address = row['pool_address']
    timeframe = 'day'
    
    url = base_url.format(network=network, pool_address=pool_address, timeframe=timeframe)
    
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        data = response.json()['data']['attributes']['ohlcv_list']
        
        ohlcv_df = pd.DataFrame(data, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
        ohlcv_df['date'] = pd.to_datetime(ohlcv_df['timestamp'], unit='s').dt.date
        
        ohlcv_df['network'] = network
        ohlcv_df['pool_address'] = pool_address
        
        final_df = ohlcv_df[['network', 'pool_address', 'date', 'open', 'high', 'low', 'close', 'volume']]
        
        file_name = f"{network}_{pool_address}.csv"
        final_df.to_csv(os.path.join(output_dir, file_name), index=False)
    else:
        print(f"Error fetching data for {network}_{pool_address}: {response.status_code}")
```
